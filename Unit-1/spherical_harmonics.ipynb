Importing necessary dependencies

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.special import sph_harm
```

Creating a spherical Coordinates theta and phi mesh-grid using numpy

```python
theta, phi = np.mgrid[0:np.pi:50j, 0:2 * np.pi:50j]  # pass 1-D array, returns 2-D arrays for each 1-D array vector
# here we pass theta and phi as two spherical coordinates array 
print(theta)
print(phi)
```

Computing the spherical harmonics for given values of quantum numbers l and m, and given coordinate matrices theta and phi

```python
l = 0
m = 0
psi = sph_harm(l,m,theta, phi)  #sph_harm function of scipy takes l, m quantum numbers as parameters and as well as the theta and phi arrays
```

Computing the probability density of the Spherical harmonics: 

```python
p_den = np.abs(psi) ** 2   # probability density is the square of the absolute value of the spherical harmonics function
```

Convert the spherical coordinates to cartesian coordinate system for given value of theta and phi where radius(r) = 1:

![](../images/spherical_coordinate.jpg)

```python

x = p_den * np.sin(theta) * np.cos(phi)
y = p_den * np.sin(theta) * np.sin(phi)
z = p_den * np.cos(theta)
```

Creating plot:


```python
fig, ax = plt.subplots(subplot_kw={"projection" : "3d"})   #creating figure and axes using subplots function

# plotting surface 
ax.plot_surface(x, y, z, rstride=3, cstride=3, facecolors=plt.cm.viridis(p_den), linewidth = 0, antialiased = False)
ax.set_title(f"Spherical Harmonics l = {l}, m = {m}")
plt.show()
```
